### Here we will be retrieving the names of the chemicals in our basis set in order to further reduce some options.

```python
import pubchempy as pcp
import pandas as pd
import numpy as np
import requests
import json
```

```python
hba = pd.read_csv('HBA_basis_set.csv')
```

```python
hba
```

```python
hbd = pd.read_csv('HBD_basis_set.csv')
```

```python
hbd
```

```python
hba_name_list = [] #empty list to append compound names too

for i, row in hba.iterrows():
    #general pubchempy request to obtain a json of the entire compound data
    cid = row['HBA_cid']
    names_request = "https://pubchem.ncbi.nlm.nih.gov/rest/pug_view/data/compound/%s/JSON" % str(cid) 
    request = requests.get(names_request)
    request_json = request.json()
    
    names = request_json['Record']['RecordTitle'] #the compound name is contained at the very top of the json, fairly easy.
    
    hba_name_list.append(names)
    
hba['HBA'] = hba_name_list
```

```python
hba
```

```python
hba.to_csv('HBA_basis_set_w_names.csv', index = False)
```
