```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from altair import *
```

```python
conductivity = pd.read_csv('conductivity/conductivity.csv')
```

```python
conductivity
```

```python
conductivity['HBD_Mole_Fraction'] = 1-conductivity['HBA Mole Fraction']
```

```python
conductivity['Ave_MW'] = (conductivity['HBD_Mole_Fraction']*conductivity['HBD_MW']) + (conductivity['HBA Mole Fraction']*conductivity['HBA_MW'])
```

```python
conductivity
```

```python
fig, ax = plt.subplots()
im = ax.scatter(conductivity['Ave_MW'], conductivity['Electrical conductivity, S/m'], c= conductivity['Temperature, K'], cmap=plt.cm.jet)

# Add a colorbar
cbar = fig.colorbar(im, ax=ax)
cbar.set_label('Temperature, K', fontsize = 12, weight = 'bold')

ax.set_xlabel('Average MW, g/mol', fontsize = 12, weight = 'bold')
ax.set_ylabel('Electrical conductivity, S/m', fontsize = 12, weight = 'bold')

fig.savefig('cond.png')


# set the color limits - not necessary here, but good to know how.
#im.set_clim(0.0, 1.0)
```

```python
density = pd.read_csv('density/density.csv')
```

```python
density
```

```python
density['HBD_Mole_Fraction'] = 1-density['HBA Mole Fraction']
```

```python
density['Ave_MW'] = (density['HBD_Mole_Fraction']*density['HBD_MW']) + (density['HBA Mole Fraction']*density['HBA_MW'])
```

```python
density
```

```python
for i, row in density.iterrows():
    if row['HBD'] == 'water':
        density = density.drop(i)
```

```python
density
```

```python
fig, ax = plt.subplots()
im = ax.scatter(density['Ave_MW'], density['Specific density, kg/m<SUP>3</SUP>'], c= density['Temperature, K'], cmap=plt.cm.jet)

# Add a colorbar
cbar = fig.colorbar(im, ax=ax)
cbar.set_label('Temperature, K', fontsize = 12, weight = 'bold')

ax.set_xlabel('Average MW, g/mol', fontsize = 12, weight = 'bold')
ax.set_ylabel('Density, kg/m^3', fontsize = 12, weight = 'bold')

fig.savefig('Dens1.png')
```
